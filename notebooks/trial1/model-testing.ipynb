```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
#reloading our prepared data

X_train_bal = pd.read_csv("../data/X_train_bal.csv")
Y_train_bal = pd.read_csv("../data/Y_train_bal.csv")
X_test = pd.read_csv("../data/X_test.csv")
Y_test = pd.read_csv("../data/Y_test.csv")
```

what all the metrics we test will mean?

1. Precision: of all the customers predicted as churners, how many of them were actually churners
2. Recall: of all the actual churners, how many did the model correctly identify
3. F1-score: The balance between precision and recall
4. Support: number of samples for that class in our test dataset

```python
# Logistic Regression

log_reg = LogisticRegression(max_iter = 1000) #creating the model
log_reg.fit(X_train_bal, Y_train_bal) #training the model (.fit() means learn from the dataset)
y_pred_lr = log_reg.predict(X_test) #make predictions
print("Logistic Regression Accuracy: ", accuracy_score(Y_test, y_pred_lr)) #evaluate the model's performance
print(classification_report(Y_test, y_pred_lr)) #metrics report
```

what the above output means??

with the help of logistic regression model, we get 53.87% accuracy only. 
This means our model isn't performing that well.

from the above table we can see that the model is able to identify churners (recall = 0.77) more 
accurately than identifying non-churners.

what's the use of macro avg and weighted avg?
macro average: treats both classes equally (this is useful when the classes are balanced)
weighted average: weights by number of samples. (useful when one class dominates)

from the above result we can come to a conclusion that Logistic Regression might be too simple for our data, since our data might be complex or huge. So we need to test our data on other models.

```python
# Random Forest

rf = RandomForestClassifier(n_estimators=200, random_state=42) #creating the model
rf.fit(X_train_bal, Y_train_bal) #training the model
y_pred_rf = rf.predict(X_test) #making predictions
print("Random Forest Accuracy: ", accuracy_score(Y_test, y_pred_rf)) #evaluating model's performance
print(classification_report(Y_test, y_pred_rf)) #metrics report
```

what the above output means??
- with the help of random forest, we get 54.42% accuracy only. This shows the model isn't performing well.
- also the model is able to identify churners and non-churners by the same percentage

```python
# XGBoost

xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42) #creating model
xgb.fit(X_train_bal, Y_train_bal) #training the model
y_pred_xgb = xgb.predict(X_test) #making predictions
print("XGBoost Accuracy: ", accuracy_score(Y_test, y_pred_xgb)) #evaluating model's performance
print(classification_report(Y_test, y_pred_xgb)) #metrics report
```

```python
#comparing results of various models

results = pd.DataFrame({
    'Model' : ['Logistic Regression', 'Random Forest', 'XGBoost'],
    'Accuracy' : [accuracy_score(Y_test, y_pred_lr),
        accuracy_score(Y_test, y_pred_rf),
        accuracy_score(Y_test, y_pred_xgb)]})

print(results)
```
