```python
# importing the required libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
import os
import joblib
from ripser import Rips
from persim import PersistenceImager
```

```python
# importing preprocessed data
combined_df = pd.read_csv("../data/combined_cleaned_encoded.csv")
```

```python
# Scaling 
# Select numerical columns for scaling
num_cols = ['age', 'tenure', 'monthlycharges']

# Initialize the scaler
scaler = StandardScaler()

# Fit the scaler on numeric columns and transform them
df_scaled = combined_df.copy()
df_scaled[num_cols] = scaler.fit_transform(combined_df[num_cols])

print(df_scaled.head())
```

```python
# TDA - Persistent Homology

# drop customer_id - we only need numeric columns 
df_numeric = df_scaled.drop(columns=['customer_id'], errors='ignore')

# Initialize Ripser
rips = Rips(maxdim=1)  # compute 0D (clusters) and 1D (loops)
diagrams = rips.fit_transform(df_numeric)
```

```python
# Initialize persistence imager
pimgr = PersistenceImager(pixel_size=0.1, spread=0.1, verbose=False)
pimgr.fit(diagrams)  # learn min/max birth/death

# Transform diagrams into images
tda_features = []
for dgm in diagrams:
    img = pimgr.transform(dgm)
    tda_features.append(img.flatten())  # flatten 2D image → 1D vector

# Combine into a numpy array
tda_features = np.array(tda_features)
print("TDA features shape:", tda_features.shape)
```

```python
# Combine TDA features with original dataset
df_tda = pd.DataFrame(tda_features, index=df_scaled.index)
df_final = pd.concat([df_scaled.reset_index(drop=True), df_tda.reset_index(drop=True)], axis=1)

print("Final feature shape:", df_final.shape)
```
