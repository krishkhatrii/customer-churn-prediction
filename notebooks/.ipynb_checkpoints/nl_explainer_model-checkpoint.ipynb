```python
import shap
import numpy as np
import joblib
```

```python
import sys
import os

# Move up one level: notebooks → project root
project_root = os.path.abspath("..")
sys.path.append(project_root)

print(project_root)
```

```python
from backend.nl_explainer import NLExplainer

feature_names = [
    "gender",
    "paymentmethod",
    "industry",
    "age",
    "tenure",
    "monthlycharges"
]

model = joblib.load("../models/super_ensemble.pkl")
mappings = joblib.load("../models/mappings.pkl")

gender_map   = {v: k for k, v in mappings["gender"].items()}
payment_map  = {v: k for k, v in mappings["paymentmethod"].items()}
industry_map = {v: k for k, v in mappings["industry"].items()}

nl_explainer = NLExplainer(
    model,
    feature_names,
    gender_map,
    payment_map,
    industry_map
)

joblib.dump(nl_explainer, "../models/nl_explainer.pkl")
```

```python
# Adjust path to your saved model
model = joblib.load("../backend/artifacts/super_ensemble.pkl")
```

```python
# VotingClassifier has named_estimators_ attribute
cat_model = model.named_estimators_["cat"]

# Optional: check type
print(cat_model)
```

```python

cat_explainer = shap.TreeExplainer(cat_model)
```

```python
# Example input vector (6 features for NLExplainer)
import numpy as np

# Example feature vector in the order: gender, paymentmethod, industry, age, tenure, monthlycharges
test_vector = np.array([1, 2, 0, 35, 12, 99]).reshape(1, -1)

# Get SHAP values
shap_values = cat_explainer.shap_values(test_vector)
print("SHAP values:", shap_values)
```
